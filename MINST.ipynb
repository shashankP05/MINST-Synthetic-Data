```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt

# Hyperparameters
latent_dim = 128
batch_size = 128
lr_G = 0.0001  # Generator learning rate (slower)
lr_D = 0.0003  # Discriminator learning rate (faster)
epochs = 150
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Check if GPU is being used
print(f"Using device: {device}")
if device.type == 'cuda':
    print(f"GPU Name: {torch.cuda.get_device_name(0)}")
    print(f"GPU Memory: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f} GB")
else:
    print("GPU not available, using CPU")
```

```python


# MNIST data loader
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([0.5], [0.5])
])
dataloader = DataLoader(
    datasets.MNIST('.', download=True, train=True, transform=transform),
    batch_size=batch_size, shuffle=True
)

# Generator - Improved architecture
class Generator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Linear(latent_dim, 256),
            nn.BatchNorm1d(256),
            nn.ReLU(True),
            nn.Linear(256, 512),
            nn.BatchNorm1d(512), 
            nn.ReLU(True),
            nn.Linear(512, 1024),
            nn.BatchNorm1d(1024),
            nn.ReLU(True),
            nn.Linear(1024, 784),
            nn.Tanh()
        )
    def forward(self, z):
        return self.model(z).view(-1, 1, 28, 28)

# Discriminator - Improved architecture
class Discriminator(nn.Module):
    def __init__(self):
        super().__init__()
        self.model = nn.Sequential(
            nn.Flatten(),
            nn.Linear(784, 512),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(512, 256),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(256, 128),
            nn.LeakyReLU(0.2),
            nn.Dropout(0.3),
            nn.Linear(128, 1),
            nn.Sigmoid()
        )
    def forward(self, x):
        return self.model(x)

# Initialize models
G = Generator().to(device)
D = Discriminator().to(device)

# Loss and optimizers with different learning rates
criterion = nn.BCELoss()
optimizer_G = optim.Adam(G.parameters(), lr=lr_G, betas=(0.5, 0.999))
optimizer_D = optim.Adam(D.parameters(), lr=lr_D, betas=(0.5, 0.999))

# Training loop with improvements
print("Starting improved GAN training...")
for epoch in range(epochs):
    for i, (real_imgs, _) in enumerate(dataloader):
        real_imgs = real_imgs.to(device)
        batch_size = real_imgs.size(0)

        # Label smoothing for better training
        real_labels = torch.ones(batch_size, 1).to(device) * 0.9  # Smooth real labels
        fake_labels = torch.zeros(batch_size, 1).to(device) + 0.1  # Smooth fake labels

        # Add small noise to real images for better stability
        noise = torch.randn_like(real_imgs) * 0.05
        real_imgs_noisy = real_imgs + noise

        # ---------------------
        #  Train Discriminator
        # ---------------------
        z = torch.randn(batch_size, latent_dim).to(device)
        fake_imgs = G(z).detach()
        
        # Real and fake losses
        real_loss = criterion(D(real_imgs_noisy), real_labels)
        fake_loss = criterion(D(fake_imgs), fake_labels)
        loss_D = (real_loss + fake_loss) / 2

        optimizer_D.zero_grad()
        loss_D.backward()
        optimizer_D.step()

        # -----------------
        #  Train Generator
        # -----------------
        z = torch.randn(batch_size, latent_dim).to(device)
        gen_imgs = G(z)
        loss_G = criterion(D(gen_imgs), real_labels)  # Want discriminator to think fakes are real

        optimizer_G.zero_grad()
        loss_G.backward()
        optimizer_G.step()
        
        # Print progress every 200 batches
        if i % 200 == 0:
            print(f"Epoch [{epoch+1}/{epochs}], Batch [{i}/{len(dataloader)}], "
                  f"Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}")

    # Print epoch summary
    print(f"Epoch [{epoch+1}/{epochs}] completed - Loss D: {loss_D.item():.4f}, Loss G: {loss_G.item():.4f}")
    
    # Generate sample images every 20 epochs to monitor progress
    if (epoch + 1) % 20 == 0:
        G.eval()
        with torch.no_grad():
            z_sample = torch.randn(16, latent_dim).to(device)
            sample_imgs = G(z_sample).cpu()
            
            fig, axs = plt.subplots(4, 4, figsize=(6, 6))
            fig.suptitle(f'Generated Images - Epoch {epoch+1}', fontsize=14)
            for j in range(4):
                for k in range(4):
                    axs[j, k].imshow(sample_imgs[j*4 + k].squeeze(), cmap="gray")
                    axs[j, k].axis('off')
            plt.tight_layout()
            plt.show()
        G.train()

print("Training completed!")

# Generate and plot final samples
z = torch.randn(16, latent_dim).to(device)
samples = G(z).cpu().detach()

fig, axs = plt.subplots(4, 4, figsize=(5, 5))
for i in range(4):
    for j in range(4):
        axs[i, j].imshow(samples[i*4 + j].squeeze(), cmap="gray")
        axs[i, j].axis('off')
plt.tight_layout()
plt.show()
```
